```python
## Getting df's of positions
df_ST = df[df['Position'] == 'ST'].count()
df_LW = df[df['Position'] == 'LW'].count()
df_RW = df[df['Position'] == 'RW'].count()
df_RF = df[df['Position'] == 'RF'].count()
df_CAM = df[df['Position'] == 'CAM'].count()
df_CM = df[df['Position'] == 'CM'].count()
df_CDM = df[df['Position'] == 'CDM'].count()
df_RB = df[df['Position'] == 'RB'].count()
df_LB = df[df['Position'] == 'LB'].count()
df_CB = df[df['Position'] == 'CB'].count()
df_GK = df[df['Position'] == 'GK'].count()
df_RM = df[df['Position'] == 'RM'].count()
df_LM = df[df['Position'] == 'LM'].count()
df_LF = df[df['Position'] == 'LF'].count()
df_CF = df[df['Position'] == 'CF'].count()
df_RWB = df[df['Position'] == 'RWB'].count()
df_LWB = df[df['Position'] == 'LWB'].count()

rm = df_RM[0]
lm = df_LM[0]
gk = df_GK[0]
rb = df_RB[0]
cb = df_CB[0]
lb = df_LB[0]
cdm = df_CDM[0]
cm = df_CM[0]
cam = df_CAM[0]
st = df_ST[0]
rw = df_RW[0]
lw = df_LW[0]
rf = df_RF[0]
lf = df_LF[0]
cf = df_CF[0]
lwb = df_LWB[0]
rwb = df_RWB[0]
```

```python
#bar plot of positions counts
players_count = [gk,rb,cb,lb,cdm,cm,cam,st,rw,lw,rf,rm,lm,lf,cf,lwb,rwb]
players = ['gk','rb','cb','lb','cdm','cm','cam','st','rw','lw','rf','rm','lm','lf','cf','lwb','rwb']
plt.figure(figsize=(8,5),dpi=300)        
plt.bar(players, players_count)
plt.title("Number of players by positions")
plt.show()
```

```python
u_50 = 0
u_40 = 0
u_35 = 0
u_30 = 0 
u_25 = 0
u_20 = 0 

for i in df.Age:
    if i >= 40:
        u_50 += 1
    elif i >= 35 and i < 40:
        u_40 += 1
    elif i >= 30 and i < 35:
        u_35 += 1
    elif i >= 25 and i < 30:
        u_30 += 1
    elif i >= 20 and i < 25:
        u_25 += 1    
    else:
        u_20 += 1

plt.figure(figsize=(8,5),dpi=300)        
ages = ['under 20', '20 - 25', '25 - 30', '30 - 35', '35 - 40', 'over 40']
values = [u_20 ,u_25, u_30, u_35, u_40, u_50]
plt.bar(ages, values)
plt.title('Number of players in age groups')
plt.show()
```
